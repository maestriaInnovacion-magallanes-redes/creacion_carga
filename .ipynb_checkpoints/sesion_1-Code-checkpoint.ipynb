```python
# link a google sheets entre "comillas" 
LinkToGoogle="https://docs.google.com/spreadsheets/d/e/2PACX-1vReqFHgPvGpZiSozcooJkZOH7kOCvtpgtq0fvuIehQ6MfqM-To93umwRY0pJBZzoCetj_ptmLCpNnGv/pub?output=xlsx"
```

Leamos la matriz de adyacencia en Python, para lo cual primero leemos la tabla usando el link en con la libreria _pandas_:

```python
# leyendo matriz de adyacencia

import pandas as pd

adjacency=pd.read_excel(LinkToGoogle, # link a la tabla
                        index_col=0,  # la columna '0' guarda los nombres
                        sheet_name='adjacency') # la tabla tiene este nombre
```

El objeto **adjancency** guarda la tabla. Ahora usamos __from_pandas_adjacency()__ de la librería **networkx**, y se crea la red:

```python
# crear red
eliteFromAdjacency=nx.from_pandas_adjacency(adjacency)
```

La red ya está creada, veamos:

```python
# nodos de la red
eliteFromAdjacency.nodes()
```

```python
# enlaces de la red
eliteFromAdjacency.edges()
```

Si quisieramos usar el _edgelist_, necesitamos:

```python
edgeList=pd.read_excel(LinkToGoogle,
                        sheet_name='edgelist') # la tabla tiene este nombre

edgeList
# crear red
eliteFromEdgelist=nx.from_pandas_edgelist(edgeList,edge_attr=True)
```

Veamos la diferencia eentre ambas redes:

* Cantidad de Nodos:

```python
# 'len()' se usa para contar elementos
eliteFromAdjacency.number_of_nodes(), eliteFromEdgelist.number_of_nodes()
```

* Cantidad de enlaces:

```python
eliteFromAdjacency.number_of_edges(), eliteFromEdgelist.number_of_edges()
```

* Las **islas** del _edgelist_:

```python
list(nx.isolates(eliteFromEdgelist))
```

* Las islas del _adjacency_:

```python
list(nx.isolates(eliteFromAdjacency))
```

La red **eliteFromEdgelist** puede completarse:

```python
eliteFromEdgelist.add_nodes_from(['Lucioni', 'Rodriguez Rodríguez', 'Custer', 'Ikeda', 'Cogorno', 'Arias Davila'])
```

Ahora tiene los mismos nodos:

```python
eliteFromEdgelist.number_of_nodes()
```

Y claro, sus islas:

```python
list(nx.isolates(eliteFromEdgelist))
```

### Exportación

Las redes creadas con Networkx pueden exportarse en diferentes formatos, que serán útiles en otros programas. Aquí grabaremos en el formato GML y GRAPHML:

```python
filename1='eliteFromAdjacency.gml' # nombre incluye 'donde'
filename2='eliteFromAdjacency.graphml'

# guardando:
nx.write_gml(eliteFromAdjacency,filename1)

nx.write_graphml(eliteFromAdjacency,filename2)
```

# VISUALIZACION DE REDES



## Usando NETWORKX

Tenemos activa la red _eliteFromAdjacency_. Visualizarla con Networkx es muy sencillo:


```python
nx.draw(eliteFromAdjacency)
```

La ubicacion de los nodos sigue por lo general algun algoritmo, este es otro:

```python
pos = nx.spiral_layout(eliteFromAdjacency) #algoritmo spiral
nx.draw(eliteFromAdjacency, pos=pos)
```

Hay muchos _layouts_ disponibles, los de Networkx puedes revisarlos [aquí](https://networkx.org/documentation/stable/reference/drawing.html#).

## En Python es algo sencillo ver los labels

```python
pos = nx.spiral_layout(eliteFromAdjacency) # algoritmo spiral
nx.draw(eliteFromAdjacency, pos=pos, with_labels=True)
```

Pero, saber los comandos a veces toma tiempo:

```python
nx.draw(
    eliteFromAdjacency, 
    pos=pos, 
    with_labels=True, 
    node_color='lightblue',   # Cambia el color del nodo para que resalte el texto
    font_size=10,             # Ajusta el tamaño de la letra
    font_weight='bold',       # Hace la letra más legible
    node_size=500             # Agranda el nodo si el texto es muy largo
)
```

## Redes con 'edges' dirigidos



```python
red2 = nx.DiGraph()

red2.add_edges_from([(6,4),(4,5),(4,3),(5,2),(5,1),(2,3),(2,1)])

# cuales tenemos:
red2.edges()
```

En las redes dirigidas como la creada, `(6,4)` es diferente a `(4,6)`, entonces, si ambos edges existen, deben incluirse manera explícita.

```python
# La forma rápida y directa
pos = nx.spiral_layout(red2) # algoritmo spiral
nx.draw(red2, with_labels=True, node_color="skyblue", pos=pos)
```

### Ejemplo ficticio: Relaciones de asesoría en un departamento de Historia


```python
asesorias = nx.DiGraph()

# Todos son profesores del departamento
profesores = [
    "Elena Vargas", "Roberto Díaz",           # 1ª generación
    "Carlos Méndez", "Ana Ruiz", "Bruno Soto", "Laura Vega",  # 2ª generación
    "Martín Herrera", "Lucía Fernández", "Sofía Reyes", "Diego Ortega"  # 3ª
]
asesorias.add_nodes_from(profesores)

# Flecha: asesor → asesorado (dirección del legado intelectual)
asesorias.add_edges_from([
    # 1ª → 2ª generación
    ("Elena Vargas", "Carlos Méndez"),
    ("Elena Vargas", "Ana Ruiz"),
    ("Roberto Díaz", "Bruno Soto"),
    ("Roberto Díaz", "Laura Vega"),
    # 2ª → 3ª generación
    ("Carlos Méndez", "Martín Herrera"),
    ("Carlos Méndez", "Lucía Fernández"),
    ("Ana Ruiz", "Sofía Reyes"),
    ("Bruno Soto", "Diego Ortega"),
    ("Laura Vega", "Diego Ortega"),  # coasesoría
])

#veamos
print(asesorias)
```

```python
nx.draw(asesorias,with_labels=True)
```

## Redes bipartitas




### Ejemplo ficticio: Prestación de servicios profesionales




```python
from networkx.algorithms import bipartite

# 1. CREAMOS EL GRAFO DIRIGIDO BIPARTITO
red_servicios = nx.DiGraph()

# Conjunto U: profesionales (bipartite=0)
profesionales = [
    "Ana", "Bruno", "Carlos", "Diana", "Elena",
    "Federico", "Gabriela", "Hugo", "Isabel", "Javier"
]

# Conjunto W: empresas cliente (bipartite=1)
empresas = ["FinanzasCorp", "TechSecure", "InnovaDigital"]

red_servicios.add_nodes_from(profesionales, bipartite=0)
red_servicios.add_nodes_from(empresas, bipartite=1)

# EDGES DIRIGIDOS: profesional → empresa (presta servicio)
red_servicios.add_edges_from([
    # Auditores financieros → FinanzasCorp
    ("Ana", "FinanzasCorp"),
    ("Bruno", "FinanzasCorp"),
    ("Carlos", "FinanzasCorp"),
    # Carlos también asesora a TechSecure en riesgos
    ("Carlos", "TechSecure"),
    # Especialistas en seguridad → TechSecure
    ("Diana", "TechSecure"),
    ("Elena", "TechSecure"),
    ("Federico", "TechSecure"),
    ("Gabriela", "TechSecure"),
    # Gabriela también asesora a InnovaDigital en compliance
    ("Gabriela", "InnovaDigital"),
    # Desarrolladores → InnovaDigital
    ("Hugo", "InnovaDigital"),
    ("Isabel", "InnovaDigital"),
    ("Javier", "InnovaDigital"),
])

# que tenemos?
print(red_servicios,"¿Es bipartita la red?", bipartite.is_bipartite(red_servicios))
```

```python
# 2. Configurar la disposición en dos columnas (bipartite_layout)
pos = nx.bipartite_layout(red_servicios, profesionales)

nx.draw(
    red_servicios,
    pos,
    with_labels=True, 
    node_color="skyblue")
```

Aqui vemos la proyeccción de la `red_servicios` sobre los nodos de profesionales, ya es una red unipartita no dirigida, donde el edge representa `han dado servicios a la misma empresa`

```python
red_unipartita = bipartite.projected_graph(red_servicios.to_undirected(), profesionales)

# 3. Graficar solo la red unipartita

pos = nx.spring_layout(red_unipartita, seed=42)

nx.draw(
    red_unipartita,
    pos,
    with_labels=True,
    node_color="orange",
    node_size=900,
    font_weight="bold",
    font_size=12,
    edge_color="dimgray",
    width=2,
)
```
